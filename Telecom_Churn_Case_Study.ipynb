### Null Value Analysis and Handling

The steps that are deployed in Handling Nulls:

1. Find the Null Value and its percentage
2. Find if the Missing values are Meaningful or Actual missing data
    Meaningful Missing data- Data that cannot be derived from other columns
    For eg.,When the recharge Date and Amount are missing and the count of recharge is NA, 
    then the count of recharge is a meaningful missing data and it should ideally be zero since no recharge was made.
    
    Actual Missing Data - Data that can be derived from other columns
    For eg., When the recharge Date and Amount are having valid values and the count of recharge is NA,
    then the count of recharge is an actual missing data as it should ideally have a valid value since the recharge was made.
   
3. Impute Meaningful Missing with 0
4. Check for Null value percentage
5. Drop the columns with Null percentage greater than 40%
6. Actual missing values are imputed with Median for continuous columns and Mode for categorical columns

#### Deriving Churn (target variable)

Based on the problem statement, Usage-based churn will be used for deriving the target variable.

Usage-based churn: Customers who have not done any usage, either incoming or outgoing - in terms of calls, internet etc. over a period of time.


In churn prediction, we assume that there are three phases of customer lifecycle :
        
        The good phase: In this phase, the customer is happy with the service and behaves as usual. Corresponds to months 6 & 7
    
    The action phase: The customer experience starts to sore in this phase. Corresponds to month 8
    
    The churn phase: In this phase, the customer is said to have churned.
    
We will define churn based on this phase. Corresponds to month 9. Also, it is important to note that at the time of prediction (i.e. the action months), this data is not available for prediction. Thus, after tagging churn as 1/0 based on this phase, we discard all data corresponding to this phase.

Below arfe the steps to derive the "churn" variable:

1. Calculate the total incoming and outgoing minutes of usage
2. Calculate 2g and 3eg data consumption
3. Create Churn Variable( based on the non usage of calls or data in month 9 )
4. Check Churn Percentage
5. Delete all the columns belonging to month 9

```python
# Univariate analyis - numerical variables with outlier handling
def uni_num(data, column):
    
    plt.figure(figsize = [18,8])

    plt.subplot(1,2,1)
    pltname = column + ' - without handling outliers'
    plt.title(pltname)
    sns.histplot(data=data, x=column, color='purple')
    plt.xticks(rotation = 45)
    
    Q1 = hvc[column].quantile(0.25)
    Q3 = hvc[column].quantile(0.75)
    IQR = Q3 - Q1
    Min_value = (Q1 - 1.5 * IQR)
    Max_value = (Q3 + 1.5 * IQR)
    print("Min value before which outlier exist: {}".format(Min_value))
    print("Max value after which outlier exist: {}".format(Max_value))
    
    plt.subplot(1,2,2)
    pltname = column + ' - with handling outliers'
    plt.title(pltname)
    sns.histplot(data=hvc, x=data[data[column] <= Max_value][column], color='orange')
    plt.xticks(rotation = 45)

    plt.tight_layout(pad = 4)
    plt.show()
```

This method may simply distort the dist variables.

## Plotting ROC curve

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```
